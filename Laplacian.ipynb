## ðŸ§  Laplacian Edge Detection with OpenCV

In this section, we implement **Laplacian edge detection**, a second-order derivative operator used to highlight regions of rapid intensity change. Unlike first-order derivatives (like Sobel), Laplacian is **isotropic** â€” it does **not depend on direction**.

---

### ðŸ“Œ Theory

The Laplacian of an image is defined as:

\[
\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2}
\]

This operation is sensitive to noise, so it's often preceded by **Gaussian smoothing** in advanced applications (e.g., LoG: Laplacian of Gaussian).

---

### âœ… Implementation in Python using OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread("your_image.jpg", cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (512, 512))  # Optional resizing

# Apply Laplacian operator (2nd derivative)
laplacian = cv2.Laplacian(image, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)

# Plot the original and Laplacian edge map
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Laplacian Edge Map")
plt.imshow(laplacian_abs, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread("laplacian.jpeg", cv2.IMREAD_GRAYSCALE)  # optional

# Apply Laplacian operator (second-order derivative)
laplacian = cv2.Laplacian(image, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)  # Convert to 8-bit for visualization

# Plot original and Laplacian edge map
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Laplacian Edge Map")
plt.imshow(laplacian_abs, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
````
