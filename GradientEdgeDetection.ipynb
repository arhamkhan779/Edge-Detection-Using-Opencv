# üîç Gradient Edge Detection using Sobel Operator (OpenCV)

In this notebook, we'll perform **gradient-based edge detection** using the **Sobel operator** in OpenCV.

---

## üìå Objective
To detect edges in an image by computing the gradient in both X and Y directions, and combining them to obtain an edge map.

---

## üß† Key Concepts

- **Gradient**: Measures change in intensity; sharp changes indicate edges.
- **Sobel Operator**: A discrete differentiation operator that computes an approximation of the gradient.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("ForGradient.jpeg")
```

```python
grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # ‚àÇI/‚àÇx
grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
```

```python

edge_map_x = cv2.convertScaleAbs(grad_x)
edge_map_y = cv2.convertScaleAbs(grad_y)
```

```python
magnitude = cv2.magnitude(grad_x, grad_y)

edge_map = cv2.convertScaleAbs(magnitude)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 6))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Edge Map (X Gradient)")
plt.imshow(edge_map_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Edge Map (Y Gradient)")
plt.imshow(edge_map_y, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Gradient Magnitude")
plt.imshow(edge_map, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
